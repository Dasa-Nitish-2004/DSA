# preorder recursive approach

```python
class Solution(object):
    def preorderTraversal(self, root):
        if root == None:
            return []
        stack = [root]
        output = []
        while stack:
            k = stack.pop()
            output.append(k.val)
            if k.right:
                stack.append(k.right)
            if k.left:
                stack.append(k.left)
        return output           
```

# inorder reqursive approach

```python
class Solution(object):
    def inorderTraversal(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[int]
        """
        output = []
        if root == None:
            return output        
        stack = [root]
        while stack[0] != None:
            if stack[-1] == None:
                stack.pop()
                k = stack.pop()
                output.append(k.val)
                stack.append(k.right)
            else:
                stack.append(stack[-1].left)
        return output
```

# interative postOrder Traversal

```python
# two stack approach
class Solution(object):
    def postorderTraversal(self, root):
        stk1 = [root]        
        stk2 = []
        if not root:
            return stk2
        while stk1:
            k = stk1.pop()
            stk2.append(k.val)
            if k.left:
                stk1.append(k.left)
            if k.right:
                stk1.append(k.right)
        return stk2[::-1]
```

### 110. Balanced Binary Tree
check whether it is balanced or not

```python
class Solution(object):
    def isBalanced(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        def rec(root):
            if not root:
                return 0
            l = rec(root.left)
            r = rec(root.right)

            if abs(l-r)>1:
                return float("inf")
            return max(l,r)+1
        
        if rec(root) == float("inf"):
            return False
        return True 
```

### Get the max diameter in the binary tree

```python
class Solution(object):
    def diameterOfBinaryTree(self, root):
        self.output = 0
        def rec(root):
            if not root:
                return 0            
            l = rec(root.left)
            r = rec(root.right)
            self.output = max(self.output,l+r)
            return max(l,r)+1
        rec(root)
        return self.output
```

### 124. Binary Tree Maximum Path Sum

```python
class Solution(object):
    def maxPathSum(self, root):
        self.maxi = -float("inf")

        def rec(node):
            if not node:
                return 0
            # Avoid negative contributions
            l = max(0, rec(node.left))
            r = max(0, rec(node.right))
            # Max path through current node
            self.maxi = max(self.maxi, l + r + node.val)
            # Return max single-path sum going up
            return node.val + max(l, r)

        rec(root)
        return self.maxi
```

### 100. Same Tree

```python
class Solution(object):
    def isSameTree(self, p, q):
        def rec(p,q):
            if not p and not q:
                return True
            elif not p or not q or p.val != q.val :
                return False            
            return rec(p.left,q.left) and rec(p.right,q.right)
        
        return rec(p,q)
        
```

### 103. Binary Tree Zigzag Level Order Traversal

```python
class Solution(object):
    def zigzagLevelOrder(self, root):
        output = []
        if not root:
            return output
        stack = [root]
        l_r = True
        while stack:
            val = []
            n = len(stack)            
            for i in range(n):
                k = stack.pop(n-1)
                n= n-1
                val.append(k.val)
                if l_r:
                    if k.left:
                        stack.append(k.left)
                    if k.right:
                        stack.append(k.right)
                else:
                    if k.right:
                        stack.append(k.right) 
                    if k.left:
                        stack.append(k.left)
            l_r = False if l_r else True
            output.append(val)
        return output
            
```
