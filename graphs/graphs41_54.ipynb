### Bellman-Ford
as dijkstra not work for negative number this algo came.

and also not work for negative cycle. which leads to TLE

```python
#User function Template for python3

class Solution:
    def bellmanFord(self, V, edges, src):
        #code here
        
        """
        algorithm is very simple 
        just doing relaxation
        but many times and how?
        
        * n-1 times
        
        for each time relax all edges if possible 
        
        
        """
        
        # step 1 : distance array
        dist = [float("inf")] * V
        
        # step 2 : mark distance of source is 0
        dist[src] = 0
        
        # step 3 : iterate n-1 time
        for _ in range(V):
            # for each iteration relax all edges
            for u,v,wt in edges:
                if dist[u]+wt < dist[v]:
                    dist[v] = dist[u]+wt
        
        # strictly we know minimum path value get by n-1 iteration.
        # to  know about cycle we need to do one more iteration 
        # if there is chance for reduction it mean negative cycle present
        temp = dist[:]
        for u,v,wt in edges:
            if temp[u]+wt < temp[v]:
                return [-1]
                # temp[v] = temp[u]+wt
        
        
        # If a vertex is unreachable from the source, its distance should be marked as 108
        for i in range(V):
            if dist[i] == float("inf"):
                dist[i] = 10**8
        
        
        return dist
```

### Floyd Warshall
multisource problem.

by end of this algo you can find smallest distance from any point to any point

how can i detect the negative cycles in it.

yes we can if in final output if there is any number other than 0 i.e. negative number 

where i==j(src = dst) that means negative cycle exists

```python
#User function template for Python
import copy
class Solution:
	def floydWarshall(self, dist):
		#Code here
	    
	    """
	    the main idea of floywarshall is brute forcing every possible way.
	    yes for time complexity : use DP
	    but leave DP understand the as it is
	    
	    example:
	    idea is let there will be graph
	    we will maintain which is the best 
	    
	    cost of i -> j is given by
	    
	    cost[i][j] =  min(cost[i][via]+cost[via][j], cost[i][j])
	    
	    
	    """
	    
	    #step 1 define cost array
	    
	    # in this case no need to define new array and do changes of direct edges
	    # values. Aim to bring the matrix to something given input dist
   	    #  [[0, 4, 108, 5, 108], [108, 0, 1, 108, 6], [2, 108, 0, 3, 108], [108, 108, 1, 0, 2], [1, 108, 108, 4, 0]]
   	    

        n = len(dist)
        const = 10**8
        for i in range(n):
            for j in range(n):
                if dist[i][j]==const:
                    dist[i][j] = float("inf")
                
                
        # Floyd-Warshall main logic:
        for via in range(n):            # Try every node as intermediate
            for i in range(n):         # For every source
                for j in range(n):     # For every destination
                    if dist[i][via] != float('inf') and dist[via][j] != float('inf'):
                        dist[i][j] = min(dist[i][j], dist[i][via] + dist[via][j])

        for i in range(n):
            for j in range(n):
                if dist[i][j]==float("inf"):
                    dist[i][j] = const
        
        return dist
   	    
```

## City With the Smallest Number of Neighbors at a Threshold Distance

```python
#User function Template for python3

from typing import List

class Solution:
    def findCity(self, n : int, m : int, edges : List[List[int]], distanceThreshold : int) -> int:
        # code here
        
        # understand the question. here we can solve using flyod Warshall
        
        # step 1 :  prepare the matrix
        
        cost = [[0 if i == j else float("inf") for i in range(n)] for j in range(n)]
        
        # step 2 add direct given edges. Note it's bidirectional
        
        for u,v,wt in edges:
            cost[u][v] = wt
            cost[v][u] = wt
        
        # step 3 write the algorithm
        
        for via in range(n):
            for i in range(n):
                for j in range(n):
                    cost[i][j] = min(cost[i][j], cost[i][via] + cost[via][j])
        
        # step 4: find the cities with mini no:of neighbour cities. if more than one mini no:of cities return greatest lable
        city_ind = -1
        mini_neighbours = float("inf")
        # we need greatest label so we iterate from 0->n
        for i in range(n): # city
            temp = 0
            for j in range(n): # neighbour
                if cost[i][j] <= distanceThreshold:
                    temp+=1
            if mini_neighbours>=temp:
                mini_neighbours = temp
                city_ind = i
        return city_ind
                
```

## Minimum Spanning Tree

```python
from heapq import heappush as push, heappop as pop
from typing import List
class Solution:
    
    #Function to find sum of weights of edges of the Minimum Spanning Tree.
    def spanningTree(self, V: int, adj: List[List[int]]) -> int:
        #code here
        # adj format: [[[1, 5], [2, 1]], [[0, 5], [2, 3]], [[1, 3], [0, 1]]]
        # step1: initiation
        
        queue = [(0,0,-1)] #(wt,node,parent) // need to select minimum wt
        visited = [0] * V
        new_edges = []
        mst_wt = 0
        
        # step 2: while loop iteration
        while queue:
            # step 3 poping
            wt, node, parent = pop(queue)
            
            # step 4 check whether it already visited?
            if visited[node]:
                continue
            
            # make node visited
            visited[node] = 1
            
            
            #add to your mst_wt
            mst_wt += wt
            
            # add neighbours to queue
            # here node in u
            for v, adj_wt in adj[node]:
                push(queue,(adj_wt,v,node))
            
            # add to new_edges
            
            if parent != -1:
                new_edges.append((parent,node))
        
        return mst_wt
        
```

### Minimum Spanning Tree
more explaination is written in book

```python
from typing import List

class Disjoint_set:
    def __init__(self,n):
        # rank based union approach
        self.rank = [0]*(n+1) # for to work both 0 and 1 based index
        self.parent = [i for i in range(n+1)]
        
        # use any one size or rank for practice purpose i wrote 
        self.size = [1]*(n+1)
        
    def find_ultimate_parent(self, node):
        if node == self.parent[node]:
            return node
        self.parent[node] = self.find_ultimate_parent(self.parent[node])
        return self.parent[node]
    
    def rank_based_union(self, u,v):
        # union required when they are disjoint sets
        
        # if same ultimate parent no need of union
        
        ulp_u = self.find_ultimate_parent(u)
        ulp_v = self.find_ultimate_parent(v)
        
        # same ultimate parent means in same tree 
        if ulp_u == ulp_v:
            return 
        
        if self.rank[ulp_u] > self.rank[ulp_v]:
            self.parent[ulp_v] = ulp_u
        elif self.rank[ulp_u] < self.rank[ulp_v]:
            self.parent[ulp_u] = ulp_v
        else:
            self.parent[ulp_u] = ulp_v
            self.rank[ulp_u] += 1
        
        
class Solution:
    
    #Function to find sum of weights of edges of the Minimum Spanning Tree.
    def spanningTree(self, V: int, adj: List[List[int]]) -> int:
        # solving using krushkal algo
        #  in krushkal steps are
        # 1. sort the edges according to weight
        # 2. pop one by one in ascending order
        # 3. make union if they are disjoint set
        # u can why only when disjoint sets?
        # if are start from smallest if already visited means it already reached with min weight
        
        
        # step 1 : sort the weights
        # we are give adj 
        # so make edges and sort
        edges = []
        for u in range(len(adj)):
            for v,wt in adj[u]:
                edges.append((wt,u,v)) # (wt, parent, node)
        
        # sort based on first element
        edges.sort(key = lambda x: x[0])
        
        
        #  we need a disjoint set so lets create class of that and use it's functionalities
        
        dj = Disjoint_set(len(adj))
        
        # pop element by element in ascending order// u can use deque
        
        mst_wt = 0
        while edges:
            wt,u,v = edges.pop(0)
            
            if dj.find_ultimate_parent(u) == dj.find_ultimate_parent(v):
                continue
            
            dj.rank_based_union(u,v)
            mst_wt+=wt
        
        
        
        return mst_wt
```
