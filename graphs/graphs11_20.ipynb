### Undirected Graph Cycle gfg

````python
class Solution:
    def isCycle(self, V, edges):
        visited = [0] * V
        adj = {}

        # Build adjacency list
        for u, v in edges:
            adj.setdefault(u, []).append(v)
            adj.setdefault(v, []).append(u)

        # BFS for cycle detection
        def bfs(i):
            queue = [(i, -1)]
            visited[i] = 1
            while queue:
                curr, parent = queue.pop(0)
                for neighbor in adj.get(curr, []):
                    if not visited[neighbor]:
                        visited[neighbor] = 1
                        queue.append((neighbor, curr))
                    elif neighbor != parent:
                        return True
            return False

        for i in range(V):
            if not visited[i]:
                if bfs(i):
                    return True
        return False
````

### 785. Is Graph Bipartite?

````python
class Solution(object):
    def isBipartite(self, graph):
        """
        :type graph: List[List[int]]
        :rtype: bool
        """
        n = len(graph)
        color_map = [-1] * n  # -1 means unvisited

        for i in range(n):
            if color_map[i] == -1:
                # Start BFS from this unvisited node
                queue = [i]
                color_map[i] = 0  # Assign color 0

                while queue:
                    node = queue.pop(0)
                    for neighbor in graph[node]:
                        if color_map[neighbor] == -1:
                            color_map[neighbor] = 1 - color_map[node]
                            queue.append(neighbor)
                        elif color_map[neighbor] == color_map[node]:
                            return False  # Same color on both ends
        return True
````

### 802. Find Eventual Safe States

To understand **"Find Eventual Safe States"** using **DFS**, let's break down the intuition, algorithm, and a **small dry run** to make it clear.

---

## ðŸ” Problem Summary

Given a directed graph (as an adjacency list), return all the nodes from which **no cycle** is reachable. These are called **eventual safe states**.

### âœ… A node is *safe* if:

* Starting from it, **every** path leads to a **terminal node** (a node with no outgoing edges), and
* It **does not** participate in any **cycle** (direct or indirect).

---

## ðŸ’¡ Intuition (DFS Approach)

We use **DFS traversal** to **detect cycles**. The idea is:

* If a node leads to a **cycle**, itâ€™s **unsafe**.
* If a node leads only to **safe** nodes, it is **safe**.

### ðŸ”„ Key Tools:

* `visited`: Tracks whether we've visited a node.
* `pathVisited`: Tracks nodes **in the current recursion path** (used to detect cycles).
* `safe`: Boolean list to mark **safe nodes**.

---

## ðŸ§  DFS Algorithm Steps:

1. For each node, do DFS.
2. If during DFS, we revisit a node in the **current DFS path** â†’ cycle exists â†’ unsafe.
3. If we finish DFS without cycles â†’ safe.
4. Memoize results to avoid rechecking.

---

## âœ… Python Code:

```python
class Solution:
    def eventualSafeNodes(self, graph):
        n = len(graph)
        visited = [0] * n
        pathVisited = [0] * n
        safe = [0] * n

        def dfs(node):
            visited[node] = 1
            pathVisited[node] = 1

            for neighbor in graph[node]:
                if not visited[neighbor]:
                    if dfs(neighbor):  # if cycle found
                        return True
                elif pathVisited[neighbor]:
                    return True  # back edge â†’ cycle

            pathVisited[node] = 0
            safe[node] = 1
            return False

        for i in range(n):
            if not visited[i]:
                dfs(i)

        return [i for i in range(n) if safe[i] == 1]
```

````python
class Solution:
    def eventualSafeNodes(self, graph):
        n = len(graph)
        visited = [0] * n
        pathVisited = [0] * n
        safe = [0] * n

        def dfs(node):
            visited[node] = 1
            pathVisited[node] = 1

            for neighbor in graph[node]:
                if not visited[neighbor]:
                    if dfs(neighbor):  # if cycle found
                        return True
                elif pathVisited[neighbor]:
                    return True  # back edge â†’ cycle

            pathVisited[node] = 0
            safe[node] = 1
            return False

        for i in range(n):
            if not visited[i]:
                dfs(i)

        return [i for i in range(n) if safe[i] == 1]
````

### Shortest path in Directed Acyclic Graph
approach is 

1. find the topological sort

2. create a dist array. Traverse to topological sort. before that make src as 0. means in this problem case explicitly mentioned source always will be 0. So,dist[0] =0

3. traverse the topological sort and update the distances while maintain minimum
   

````python
from typing import List

class Solution:
    def shortestPath(self, V, E, edges):
        # Step 1: Build the adjacency list
        adj = {i: [] for i in range(V)}
        for u, v, wt in edges:
            adj[u].append((v, wt))

        # Step 2: Topological Sort using DFS
        visited = [0] * V
        topo_stack = []

        def dfs(node):
            visited[node] = 1
            for nei, _ in adj.get(node, []):
                if not visited[nei]:
                    dfs(nei)
            topo_stack.append(node)  # post-order

        for i in range(V):
            if not visited[i]:
                dfs(i)

        topo_stack.reverse()

        # Step 3: Initialize distances
        dist = [float('inf')] * V
        dist[0] = 0  # assuming source is node 0

        # Step 4: Relaxation along topological order
        for u in topo_stack:
            if dist[u] != float('inf'):
                for v, wt in adj[u]:
                    # maintain minimum distance
                    if dist[v] > dist[u] + wt:
                        dist[v] = dist[u] + wt

        # Convert unreachable distances from inf to -1
        for i in range(V):
            if dist[i] == float('inf'):
                dist[i] = -1

        return dist
````

### Shortest Path in Undirected Graph

````python
class Solution:
    def shortestPath(self, adj, src):
        # code here
        V = len(adj)
        dist = [-1]*V
        
        dist[src] = 0
        
        # lets apply bfs
        
        queue = [(src,0)]
        while queue:
            for _ in range(len(queue)):
                ele,dst = queue.pop(0)
                for i in adj[ele]:
                    if dist[i] == -1:
                        queue.append((i,dst+1))
                        dist[i] = dst+1
                    else:
                        dist[i] = min(dist[i],dst+1)
        return dist
````

### 127. Word Ladder leetcode

````python
class Solution(object):
    def ladderLength(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: int
        """
        alpha = "abcdefghijklmnopqrstuvwxyz"

        # Convert the word list into a set of tuples for O(1) lookups
        wordList = {tuple(word) for word in wordList}

        # Convert endWord to a tuple for comparison
        endWord = tuple(endWord)

        # Initialize the queue for BFS: (current_word_as_list, current_distance)
        queue = [(list(beginWord), 1)]  # Start at 1 to include beginWord in path length

        while queue:
            for _ in range(len(queue)):
                word, count = queue.pop(0)

                # If we reach the endWord, return the path length
                if tuple(word) == endWord:
                    return count

                # Try all possible one-letter transformations
                for i in range(len(word)):
                    original_char = word[i]
                    for ch in alpha:
                        if ch == original_char:
                            continue
                        new_word = word[:]        # Make a copy to avoid modifying the original
                        new_word[i] = ch          # Change one character
                        new_tuple = tuple(new_word)

                        # If the new word is in the word list, it's a valid transformation
                        if new_tuple in wordList:
                            wordList.remove(new_tuple)  # Mark as visited by removing
                            queue.append((new_word, count + 1))  # Add to BFS queue with incremented level

        # If no path was found
        return 0
````
