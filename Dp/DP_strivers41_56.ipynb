### 1048. Longest String Chain

````python
class Solution:
    def longestStrChain(self, words: List[str]) -> int:
        words.sort(key=len)
        
        def comp(a,b):
            if len(a) != len(b)+1:
                return False
            p1 = 0
            p2 = 0
            count=0
            while p1 < len(a) or p2 < len(b): 
                if p2 < len(b) and a[p1] == b[p2]:
                    p1+=1
                    p2+=1
                else:
                    p1+=1
                    count+=1
                    if count > 1:
                        return False
            return True


        n = len(words)
        if n == 1:
            return 1
        dp = [1]*n
        maxi = 0
        for i in range(1,n):
            for j in range(i):
                if comp(words[i],words[j]) and dp[i] < (dp[j]+1):
                    dp[i] = dp[j]+1
            maxi = max(maxi,dp[i])

        return maxi
````

````python

from typing import List


class Solution:
    def LongestBitonicSequence(self, n : int, nums : List[int]) -> int:
        # code here
        n = len(nums)
        dp1 = [1]*n
        dp2 = [1]*n
        for i in range(1,n):
            for j in range(i):
                if nums[i] > nums[j] and dp1[i] < dp1[j]+1:
                    dp1[i] = dp1[j]+1
        for i in range(n-2,-1,-1):
            for j in range(n-1,i,-1):
                if nums[i] > nums[j] and dp2[i] < dp2[j]+1:
                    dp2[i] = dp2[j]+1
        maxi = 0
        for i in range(n):
            # use this condition to avoid strictly increasing or strictly decreasing
            if dp1[i] == 1 or dp2[i] == 1:
                continue
            maxi = max(maxi,dp1[i]+dp2[i]-1)
        return maxi
                
        
````

### 673. Number of Longest Increasing Subsequence

Great! Let's walk through the example step by step using your function:

---

## 🎯 Input:

```python
nums = [1, 3, 5, 4, 7]
```

We will compute two arrays:

* `dp[i]`: Length of the **longest increasing subsequence** ending at index `i`
* `cp[i]`: **Number of such LIS** ending at index `i`

---

## 🔢 Initialization:

We start with:

```python
dp = [1, 1, 1, 1, 1]  # every element can at least form a subsequence of length 1
cp = [1, 1, 1, 1, 1]  # and there is 1 such sequence initially
```

---

## 🔄 Step-by-Step Table Fill:

### 🔁 i = 0:

No `j < i`, so skip.

---

### 🔁 i = 1 (`nums[1] = 3`)

* j = 0: `nums[1] > nums[0]` → `3 > 1`

  * `dp[1] = dp[0] + 1 = 2`
  * `cp[1] = cp[0] = 1`

```python
dp = [1, 2, 1, 1, 1]
cp = [1, 1, 1, 1, 1]
```

---

### 🔁 i = 2 (`nums[2] = 5`)

* j = 0: `5 > 1` → `dp[2] = 2`, `cp[2] = 1`
* j = 1: `5 > 3` → `dp[2] = 3`, `cp[2] = 1` (overwrite)

```python
dp = [1, 2, 3, 1, 1]
cp = [1, 1, 1, 1, 1]
```

---

### 🔁 i = 3 (`nums[3] = 4`)

* j = 0: `4 > 1` → `dp[3] = 2`, `cp[3] = 1`
* j = 1: `4 > 3` → `dp[3] = 3`, `cp[3] = 1`
* j = 2: `4 < 5` → skip

```python
dp = [1, 2, 3, 3, 1]
cp = [1, 1, 1, 1, 1]
```

---

### 🔁 i = 4 (`nums[4] = 7`)

* j = 0: `7 > 1` → `dp[4] = 2`, `cp[4] = 1`
* j = 1: `7 > 3` → `dp[4] = 3`, `cp[4] = 1`
* j = 2: `7 > 5` → `dp[4] = 4`, `cp[4] = 1`
* j = 3: `7 > 4`, and `dp[3]+1 == dp[4]` → another LIS of same length

  * `cp[4] += cp[3] = 1 + 1 = 2`

```python
dp = [1, 2, 3, 3, 4]
cp = [1, 1, 1, 1, 2]
```

---

## 🧮 Final Result:

* Max length = `4`
* Indices where `dp[i] == 4`: only `i = 4`
* Sum of `cp[i]` where `dp[i] == 4`: `cp[4] = 2`

---

### ✅ Final Output: `2`

---

Would you like a diagram to visualize how these LIS branches form?


````python
class Solution:
    def findNumberOfLIS(self, nums: List[int]) -> int:
        n = len(nums)
        if n == 1:
            return 1
        dp = [1]*n
        cp = [1]*n
        ans = 0
        maxi = 0
        for i in range(1,n):
            for j in range(i):
                if nums[i] > nums[j]:
                    if dp[i] < dp[j]+1:
                        dp[i] = dp[j]+1
                        cp[i] = cp[j]
                    elif dp[i] == dp[j]+1:
                        cp[i] += cp[j]
                    
            if dp[i] > maxi:
                ans = i
                maxi = dp[i]
        return sum(cp[i] for i in range(n) if dp[i] == maxi)
````

### Matrix Chain Multiplication
Given an array arr[] which represents the dimensions of a sequence of matrices where the ith matrix has the dimensions 

(arr[i-1] x arr[i]) for i>=1, 

find the most efficient way to multiply these matrices together. 

The efficient way is the one that involves the least number of multiplications.

````python
class Solution:
    def matrixMultiplication(self, arr):
        # code here
        def rec(i,j):
            if i == j:
                return 0
            if dp[i][j] != -1:
                return dp[i][j]
            mini = float("inf")
            for k in range(i,j):
                steps = arr[i-1]*arr[k]*arr[j]+rec(i,k)+rec(k+1,j)
                mini = min(mini,steps)
            dp[i][j] = mini
            return mini
            
        n = len(arr)
        dp = [[-1 for _ in range(n)] for _ in range(n)]
        
        return rec(1,len(arr)-1)
        
        
````

132. Palindrome Partitioning II

Given a string s, partition s such that every substring of the partition is a palindrome.

Return the minimum cuts needed for a palindrome partitioning of s.

````python
from functools import lru_cache
class Solution:
    def minCut(self, s: str) -> int:
        def isPalindrome(i,j):
            if i==j:
                return True
            p = i
            q = j
            while p<q:
                if s[p] == s[q]:
                    p+=1
                    q-=1
                else:
                    return False
            return True
        @lru_cache(None)
        def rec(i):
            if i == n:
                return 0
            mini = float("inf")
            for j in range(i,n):
                if isPalindrome(i,j):
                    cost = 1 + rec(j+1)
                    mini = min(mini,cost)
            return mini

        n = len(s)
        return rec(0)-1
````
